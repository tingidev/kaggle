### Descriptive variable: Item Description

```python
stopWords = set(stopwords.words('english'))
def wordCount(text):
    try:
        # Convert to lower case and strip regex
        text = text.lower()
        regex = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n]')
        txt = regex.sub(" ", text)
        # Remove stop words
        words = [w for w in txt.split(" ") \
                if not w in stopWords and len(w)>3]
        return len(words)
    except:
        return 0
```

```python
train['desc_len'] = train['item_description'].apply(lambda x: wordCount(x))
test['desc_len'] = test['item_description'].apply(lambda x: wordCount(x))
```

### Text processing - Item description

```python
string.punctuation
```

```python
# Tokenize
stopWords = set(stopwords.words('english'))
def tokenize(text):
    """
    sent_tokenize(): breaks text into sentences
    word_tokenize(): breaks sentences into words
    """
    try:
        # Remove punctuation
        regex = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n]')
        text = regex.sub(" ", text)
        
        # Tokenize
        tokens_ = word_tokenize(text)
        tokens = [w.lower() for w in tokens_ if w.lower() not in stopWords and re.search('[a-zA-Z]', w) and len(w)>=3] 
        
        return tokens
        
    except TypeError as e: print(text,e)
```

```python
# Apply tokenizer function to item description
train['tokens'] = train['item_description'].map(tokenize)
test['tokens'] = test['item_description'].map(tokenize)
```

```python
for description, tokens in zip(train['item_description'].head(10),train['tokens'].head(10)):
    print('Description:', description)
    print('Tokens:', tokens)
    print()
```

```python
# Build dictionary with key=cat and values as all descriptions
cat_desc = dict()
for cat in general_cats:
    text = "".join(train.loc[train['general_cat']==cat, 'item_description'].values)
    cat_desc[cat] = tokenize(text)
```

```python
# Find most common words
women100 = Counter(cat_desc['Women']).most_common(100)
beauty100 = Counter(cat_desc['Beauty']).most_common(100)
kids100 = Counter(cat_desc['Kids']).most_common(100)
electronics100 = Counter(cat_desc['Electronics']).most_common(100)
```

```python
def generate_wordcloud(tup):
    wordcloud = WordCloud(background_color='white',
                         max_words=50, max_font_size=40,
                         random_state=42
                         ).generate(str(tup))
    return wordcloud
```

```python
#Plot wordcloud of top100 words
fig,axes = plt.subplots(2, 2, figsize=(30, 15))

#Women
ax = axes[0, 0]
ax.imshow(generate_wordcloud(women100), interpolation='bilinear')
ax.axis('off')
ax.set_title('Women top 100', fontsize=30)

#Beauty
ax = axes[0, 1]
ax.imshow(generate_wordcloud(beauty100), interpolation='bilinear')
ax.axis('off')
ax.set_title('B eauty top 100', fontsize=30)

#Kids
ax = axes[1, 0]
ax.imshow(generate_wordcloud(kids100), interpolation='bilinear')
ax.axis('off')
ax.set_title('Kids top 100', fontsize=30)

#Electronics
ax = axes[1, 1]
ax.imshow(generate_wordcloud(electronics100), interpolation='bilinear')
ax.axis('off')
ax.set_title('Electronics top 100', fontsize=30)
```
